## Inventory Data

```python
with open('Resource.json') as json_file:
    Products = json.load(json_file)

cols = ['Product_ID','Name','Category','Brand','Supplier','Product Weight','Flavour','Ordered Quantity','Unit Price','Current Stock']
lst_products = pd.DataFrame(columns = cols)
for k in Products.keys():
    lst_products = lst_products.append({'Product_ID':k,
                                        'Name':Products[k]['Product_Name'],
                                        'Category':Products[k]["Category"],
                                        'Brand':Products[k]["Brand"],
                                        'Supplier':Products[k]["Supplier"],
                                        'Product Weight':Products[k]["Product_Weight"],
                                        'Flavour':Products[k]["Flavour"],
                                        'Ordered Quantity':Products[k]["Ordered_Quantity"],
                                        'Unit Price':Products[k]["Unit_Price"],
                                        'Current Stock':Products[k]["Current_Stock"]},ignore_index=True)

pd.set_option('display.max_rows', None)

display(lst_products.style.hide_index())
```

## Add New Products


```python
# Add new product records to inventory.
# To exit enter N after adding all new products.

with  open('Resource.json') as json_file:
    Products = json.load(json_file)

new_id = int(max(Products.keys()))

while True:
    new_id = new_id + 1
    name = input("Enter Product Name: ")
    category = input("Enter Category: ")
    brand = input("Enter Brand name: ")
    supplier = input("Enter Suppliers Name:")
    weight = input("Enter Products unit weight in grams: ")
    flavour = input("Enter Flavour (if not available enter NA): ")
    Ord_quantity = input("Enter Ordered Quantity: ")
    price = input("Enter Unit Price of product: ")
    
    print("\nData Saved Successfuly \n")
    Products[str(new_id)] = {'Product_Name':name,'Category':category,'Brand':brand,'Supplier':supplier,'Product_Weight':weight,'Flavour':flavour,'Unit_Price':price,'Ordered_Quantity':Ord_quantity,'Current_Stock':Ord_quantity}
    choice= input("Do you wish to Add more items (Y/N): ")
    if(choice == 'N' or choice == 'n'):
        break;

with open('Resource.json', 'w') as json_file:
    json.dump(Products, json_file)
```

## Update Inventory

```python
with  open('Resource.json') as json_file:
    Products = json.load(json_file)

pro_id = input("Enter the product id : ")

if pro_id in Products.keys():
    print("\nProduct Details.....")
    cols = ['Product_ID','Name','Category','Brand','Supplier','Product Weight','Flavour','Ordered Quantity','Unit Price','Current Stock']
    del_product = pd.DataFrame(columns=cols)
    del_product = del_product.append({'Product_ID':pro_id,
                                      'Name':Products[pro_id]['Product_Name'],
                                      'Category':Products[pro_id]["Category"],
                                      'Brand':Products[pro_id]["Brand"],
                                      'Supplier':Products[pro_id]["Supplier"],
                                      'Product Weight':Products[pro_id]["Product_Weight"],
                                      'Flavour':Products[pro_id]["Flavour"],
                                      'Ordered Quantity':Products[pro_id]["Ordered_Quantity"],
                                      'Unit Price':Products[pro_id]["Unit_Price"],
                                      'Current Stock':Products[pro_id]["Current_Stock"]},ignore_index=True)
    display(del_product.style.hide_index())
    print("\nList of properties that can be modified:")
    print("\nName , Category , Brand , Supplier , Product Weight , Flavour , Ordered Quantity , Unit Price , Current Stock \n\n")
    
    while(True):
        change_prop = input("Enter the property name you wish to change : ")
        
        if (change_prop == 'Name' or change_prop == 'name'):
            Products[pro_id]["Name"] = input("Enter New 'Name' for the product: ")
            
        elif (change_prop == 'Category' or change_prop == 'Category'):
            Products[pro_id]["Category"] = input("Enter 'Category' for the product: ")
            
        elif (change_prop == 'Brand' or change_prop == 'Brand'):
            Products[pro_id]["Brand"] = input("Enter New 'Brand' for the product: ")
            
        elif (change_prop == 'Supplier' or change_prop == 'Supplier'):
            Products[pro_id]["Supplier"] = input("Enter New Supplier for the product: ")
            
        elif (change_prop == 'Product Weight' or change_prop == 'product weight'):
            Products[pro_id]["Product Weight"] = input("Enter updated 'Weight' for the product: ")
            
        elif (change_prop == 'Flavour' or change_prop == 'flavour'):
            Products[pro_id]["Flavour"] = input("Enter updated 'Flavour' for the product: ")
            
        elif (change_prop == 'Ordered Quantity' or change_prop == 'ordered quantity'):
            Products[pro_id]["Ordered Quantity"] = input("Enter updated 'Ordered Quantity' for the product: ")
            
        elif (change_prop == 'Unit Price' or change_prop == 'unit price'):
            Products[pro_id]["Unit Price"] = input("Enter updated 'Unit Price' for the product: ")
            
        elif (change_prop == 'Current Stock' or change_prop == 'current stock'):
            Products[pro_id]["Current Stock"] = input("Enter updated 'Current' Stock for the product: ")
        else:
            print("Incorrect Choice.......")
            
        choice = input("\nDo you wish to update more propeties for this product (Y/N):")
        if choice == 'N' or choice =='n':
            with open('Resource.json', 'w') as json_file:
                json.dump(Products, json_file)
            print("\nAll changes applied to the data successfully...")
            break
else:
    print("No such product with id {} found.".format(pro_id))
    
```

## Delete Product

```python
with open('Resource.json') as json_file:
    Products = json.load(json_file)

    
id = input("Enter Product ID whose data needs to be deleted: ")

if id in Products.keys():
    cols = ['Product_ID','Name','Category','Brand','Supplier','Product Weight','Flavour','Ordered Quantity','Unit Price','Current Stock']
    del_product = pd.DataFrame(columns=cols)
    del_product = del_product.append({'Product_ID':id,
                                      'Name':Products[id]['Product_Name'],
                                      'Category':Products[id]["Category"],
                                      'Brand':Products[id]["Brand"],
                                      'Supplier':Products[id]["Supplier"],
                                      'Product Weight':Products[id]["Product_Weight"],
                                      'Flavour':Products[id]["Flavour"],
                                      'Ordered Quantity':Products[id]["Ordered_Quantity"],
                                      'Unit Price':Products[id]["Unit_Price"],
                                      'Current Stock':Products[id]["Current_Stock"]},ignore_index=True)

    print("The following data is being deleted.....\n")
    display(del_product.style.hide_index())
    
    del Products[id]
    
    with open('Resource.json', 'w') as json_file:
        json.dump(Products, json_file)
    print("Data has been deleted successfully.")
else:
    print("No product with the id - {} found in the invetory. Please check the id and try again.".format(id))
```

## New Purchases

```python
with open('Resource.json') as Resource_file:
    Products = json.load(Resource_file)

if os.path.getsize('Sales.json') == 0:
    Purchase_id = 1001

else:
    with open('Sales.json') as sales_file:
        Sales = json.load(sales_file)

    dt = max(Sales.keys())
    Purchase_id = str(int(max(Sales[dt].keys()))+1)

d = str(date.today())

purchase_list = []

Cust_Name = input("Customer Name : ")
Cust_Contact_no = input("Customer Contact Number : ")

while True:
    Id = input("\nEnter Product Id : ")
    quantity = int(input("Enter Quantity of product : "))
    
    if(quantity > int(Products[Id]['Current_Stock'])):
        print("Sorry but we only have {} {} in our stock.".format(Products[Id]['Current_Stock'],Products[Id]['Product_Name']))
        choice = input("Whould you like to buy {} {} (Y/N)".format(Products[Id]['Current_Stock'],Products[Id]['Product_Name']))
        if (choice == 'Y' or choice == 'y'):
            quantity = int(Products[Id]['Current_Stock'])
        else:
            quantity = 0
            
    if(quantity != 0):
        purchase_list.append([Id,Products[Id]['Product_Name'],quantity])
        Products[Id]['Current_Stock'] = str(int(Products[Id]['Current_Stock']) - quantity)
        
    choice = input("\nDo you want to enter more items to the list (Y/N): ")
        
    if(choice == 'Y' or choice == 'y'):
        continue
    else:
        if len(purchase_list)>0:
            bill = pd.DataFrame(columns = ['Product ID','Product Name','Quantity','Unit Price','Total'])
            
            for x in range(len(purchase_list)):
                bill = bill.append({'Product ID':purchase_list[x][0],'Product Name':purchase_list[x][1],'Quantity':purchase_list[x][2],'Unit Price':Products[purchase_list[x][0]]['Unit_Price'],'Total':int(purchase_list[x][2])*int(Products[purchase_list[x][0]]['Unit_Price'])},ignore_index=True)
            
            Total_bill =sum(bill['Total'])
            
            print("\n")
            print(tabulate(bill, headers='keys',showindex=False,tablefmt = 'psql'))
            print("Total Amount : {}".format(Total_bill))
                        
            if (Purchase_id == 1001):
                Sales = {d:{Purchase_id:{"Customer Name":Cust_Name,"Contact Number":Cust_Contact_no,"Purchases":purchase_list,"Total Bill":str(Total_bill)}}}
            else:
                if d in Sales.keys():
                    Sales[d][Purchase_id] = {"Customer Name":Cust_Name,"Contact Number":Cust_Contact_no,"Purchases":purchase_list,"Total Bill":str(Total_bill)}
                else:
                    Sales[d]={Purchase_id:{"Customer Name":Cust_Name,"Contact Number":Cust_Contact_no,"Purchases":purchase_list,"Total Bill":str(Total_bill)}}
                

            with open('Resource.json', 'w') as json_file:
                json.dump(Products, json_file)

            with open('Sales.json', 'w') as json_file:
                json.dump(Sales, json_file)
    break
```

## Billing Records

```python
with open('Sales.json') as json_file:
    Sales = json.load(json_file)

cols = ['Purchase_ID','Buyer_Name','Phone_No','Purchase_Date','Product_Id','Product_Name','Quantity','Total_Bill_Ammount']
lst_purchases = pd.DataFrame(columns = cols)

for dt in Sales.keys():
    for id in Sales[dt].keys():
        k = 0
        for product in Sales[dt][id]["Purchases"]:
            if k==0:
                purchase_id = id
                buyer = Sales[dt][id]['Customer Name']
                p_no = Sales[dt][id]['Contact Number']
                purchase_date = dt
                bill = ' '
                
            elif k == len(Sales[dt][id]["Purchases"])-1:
                purchase_id = ' '
                buyer = ' '
                p_no = ' '
                purchase_date = ' '
                bill = Sales[dt][id]['Total Bill']

            else:
                purchase_id = ' '
                buyer = ' '
                p_no = ' '
                purchase_date = ' '
                bill = ' '
            lst_purchases = lst_purchases.append({'Purchase_ID':purchase_id,
                                                  'Buyer_Name':buyer,
                                                  'Phone_No':p_no,
                                                  'Purchase_Date':purchase_date,
                                                  'Product_Id':product[0],
                                                  'Product_Name':product[1],
                                                  'Quantity':product[2],
                                                  'Total_Bill_Ammount':bill},ignore_index=True)
            k = k + 1;

pd.set_option('display.max_rows', None)

display(lst_purchases.style.hide_index())
```

## Suppliers List

```python
with open('Suppliers.json') as json_file:
    suppliers = json.load(json_file)

cols = ['Supplier ID','Company Name','Company Owner','Phone No.','Email Id','Location',]
lst_suppliers = pd.DataFrame(columns = cols)

for key in suppliers.keys():
    lst_suppliers = lst_suppliers.append({'Supplier ID':key,
                                          'Company Name':suppliers[key]['Company Name'],
                                          'Company Owner':suppliers[key]["Owners Name"],
                                          'Phone No.':suppliers[key]["Phone No"],
                                          'Email Id':suppliers[key]["Email Id"],
                                          'Location':suppliers[key]["Location"],},ignore_index=True)

pd.set_option('display.max_rows', None)

display(lst_suppliers.style.hide_index())
```
